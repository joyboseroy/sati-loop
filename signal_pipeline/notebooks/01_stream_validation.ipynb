# 01 — Stream Validation

Verify that the BrainFlow stream is working correctly
and EEG signal quality is acceptable before calibration.

```python
# Install: pip install brainflow numpy matplotlib
import numpy as np
import matplotlib.pyplot as plt
from brainflow.board_shim import BoardShim, BrainFlowInputParams, BoardIds

print("BrainFlow imported successfully")
```

```python
# Configure board
# Replace with your Muse S MAC address
params = BrainFlowInputParams()
params.mac_address = "XX:XX:XX:XX:XX:XX"
board = BoardShim(BoardIds.MUSE_S_BOARD, params)
board.prepare_session()
board.start_stream()
```

```python
# Collect 5 seconds of data
import time
time.sleep(5)
data = board.get_board_data()
print(f"Shape: {data.shape}")
print(f"EEG channels: {BoardShim.get_eeg_channels(BoardIds.MUSE_S_BOARD)}")
```

```python
# Plot EEG channels
eeg_ch = BoardShim.get_eeg_channels(BoardIds.MUSE_S_BOARD)
fig, axes = plt.subplots(len(eeg_ch), 1, figsize=(12, 8))
for i, ch in enumerate(eeg_ch):
    axes[i].plot(data[ch, :])
    axes[i].set_ylabel(f"Ch {ch}")
plt.tight_layout()
plt.show()
```

```python
# Check for obvious artifacts
import sys, os
sys.path.insert(0, os.path.join(".."))
from artifact_rejection import amplitude_reject
eeg = data[eeg_ch, :]
print(f"Amplitude artifact detected: {amplitude_reject(eeg)}")
board.stop_stream()
board.release_session()
```
