# 04 — Classifier Baseline

Train the personalised mind-wandering classifier on Phase A calibration data.

```python
import numpy as np
import sys, os
sys.path.insert(0, os.path.join(".."))
from classifier import SatiClassifier, LABEL_MAP
print("Imports OK")
```

```python
# Load Phase A labels (replace with real data path)
# Labels: 0=attentive, 1=wandering, 2=laxity
# For demonstration: generate synthetic data
np.random.seed(42)
n_per_class = 30
X = np.vstack([
    np.random.randn(n_per_class, 10) + [0.5]*10,  # attentive
    np.random.randn(n_per_class, 10) + [-0.5]*10, # wandering
    np.random.randn(n_per_class, 10) + [0]*10,    # laxity
])
y = np.array([0]*n_per_class + [1]*n_per_class + [2]*n_per_class)
print(f"Training data: {X.shape}, labels: {dict(zip(*np.unique(y, return_counts=True)))}")
```

```python
# Train classifier
clf = SatiClassifier()
clf.train(X, y)
print("Classifier trained.")

# Test prediction
sample = X[0]
probs = clf.predict_proba(sample)
for i, p in enumerate(probs):
    print(f"  P({LABEL_MAP[i]}): {p:.3f}")
```

```python
# Save classifier
os.makedirs("models", exist_ok=True)
clf.save("models/classifier_demo.pkl")
print("Saved to models/classifier_demo.pkl")

# Load and verify
clf2 = SatiClassifier.load("models/classifier_demo.pkl")
probs2 = clf2.predict_proba(sample)
print(f"Load/save round-trip verified: {np.allclose(probs, probs2)}")
```
