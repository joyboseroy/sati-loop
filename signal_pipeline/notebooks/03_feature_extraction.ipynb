```python
import numpy as np
import sys, os
sys.path.insert(0, os.path.join(".."))
from fast_features import extract_layer1_features
from slow_context import extract_somatic_context
print("Imports OK")
```

```python
# Simulate EEG window: [Fp1, Fp2, Fz, Cz]
FS = 250
N = 125
t = np.arange(N) / FS
eeg = np.stack([
    np.sin(2*np.pi*10*t) + 0.1*np.random.randn(N),  # Fp1: alpha
    np.sin(2*np.pi*10*t) + 0.1*np.random.randn(N),  # Fp2: alpha
    np.sin(2*np.pi*6*t)  + 0.1*np.random.randn(N),  # Fz: theta
    np.sin(2*np.pi*6*t)  + 0.1*np.random.randn(N),  # Cz: theta
])
features = extract_layer1_features(eeg, FS)
for k, v in features.items():
    print(f"  {k}: {v:.4f}")
```
